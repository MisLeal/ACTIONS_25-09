```python
hora = pd.read_excel('SET2025_Hora Hora - Leal Modelo_2 20.09.xlsx', sheet_name='Visão Rep.', skiprows=8) 
```

```python
hora.head()
```

```python
hora = hora[['NOME','GESTIONES','DATA', 'SUPERVISOR', 'CONTATO DIRETO']]
hora.head()
```

```python
hora.groupby(['DATA', 'NOME', 'SUPERVISOR','CONTATO DIRETO'])['GESTIONES'] \
    .sum() \
    .reset_index() \
    .sort_values(['DATA', 'GESTIONES'], ascending=[True, False])
```

```python
hora['DATA'] = pd.to_datetime(hora['DATA'], errors='coerce')
```

```python
hora = hora.dropna(subset=['DATA'])
```

```python
hora.to_csv('HORA.csv', index=False)
```
